```python
#What are the opening and closing times for a specific museum?
sql_query="SELECT museum.name, museum_hours.open, museum_hours.close  FROM museum  JOIN museum_hours ON museum.museum_id = museum_hours.museum_id  GROUP BY museum.name, museum_hours.open, museum_hours.close"
df_mysql=pd.read_sql(sql_query,connection)
df_mysql 
```

# The resulting DataFrame presents the name of each museum along with its corresponding opening and closing times.






```python
#How many artworks are available for sale in each museum?
sql_query="SELECT distinct museum.name As Museum_Name,count(work.work_id) As Artwork_for_sale FROM museum  join work  on museum.museum_id = work.museum_id group by museum.name limit 22"
df_mysql=pd.read_sql(sql_query,connection)
df_mysql 
plt.figure(figsize=(10, 10))
sns.barplot(data=df_mysql, x='Museum_Name', y='Artwork_for_sale')
plt.title('Artwork for Sale per Museum')
plt.xlabel('Museum Name')
plt.ylabel('Artwork Count')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

# The resulting bar plot visualizes this data, offering a comparative view of artwork availability across different museums, thereby helping in understanding the diversity and scale of artworks for sale within each institution.

```python
#Which artist has artworks in the most diverse styles?
sql_query="SELECT artist.full_name,work.style from artist join work on artist.artist_id=work.artist_id where work.style='diverse'"
df_mysql=pd.read_sql(sql_query,connection)
df_mysql
```

# The SQL query identifies the artist(s) whose artworks span across the most diverse range of styles, focusing specifically on artworks categorized under the style 'diverse'.






```python
# How many artworks does each artist have in a specific style?
sql_query="SELECT artist.full_name,work.style,count(work.work_id) As Artworks_count from artist join work on artist.artist_id=work.artist_id group by work.style,artist.full_name order by Artworks_count desc limit 50"
df_mysql=pd.read_sql(sql_query,connection)
df_mysql
```

# The resulting DataFrame provides insights into the prolificacy of artists within particular artistic styles, aiding in understanding their contributions and influence in those stylistic genres.

```python
#List all the artworks along with their associated museums.
sql_query="SELECT museum.name AS museum_name,work.name AS artwork_name, COUNT(*) AS artwork_count FROM museum JOIN work ON museum.museum_id = work.museum_id GROUP BY museum.name,work.name"
df_mysql=pd.read_sql(sql_query,connection)
df_mysql
```

# The resulting DataFrame presents a detailed overview of the artworks housed in each museum, facilitating an understanding of the distribution and diversity of the art collections across different institutions.






```python
#How many artworks does each museum have?
sql_query="SELECT museum.name AS museum_name,COUNT(work.name) AS artwork_count FROM museum JOIN work ON museum.museum_id = work.museum_id GROUP BY museum.name limit 22"
df_mysql=pd.read_sql(sql_query,connection)
df_mysql
plt.figure(figsize=(10, 6))
sns.barplot(data=df_mysql, x='museum_name', y='artwork_count', palette='viridis')
plt.xlabel('Museum')
plt.ylabel('Number of Artworks')
plt.title('Number of Artworks in Each Museum')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

# The SQL query tallies the number of artworks housed in each museum, providing a clear comparison of their respective art collections. The resulting bar plot visualizes this data, enabling a quick understanding of the scale and diversity of artworks present in different museums.






```python
#What is the total size (in square units) of all the artworks in a specific museum?
sql_query="SELECT museum.name AS Museum_Name,SUM(canvas_size.width * canvas_size.height) AS Total_Artwork_Size FROM museum JOIN work ON museum.museum_id = work.museum_id JOIN product_size ON work.work_id = product_size.work_id JOIN canvas_size ON product_size.size_id = canvas_size.size_id GROUP BY museum.name"
df_mysql=pd.read_sql(sql_query,connection)
df_mysql
```

# The SQL query calculates the total size of all artworks, measured in square units, within each specific museum by considering the width and height of each artwork's canvas. The resulting DataFrame provides a comprehensive overview of the collective physical footprint of artworks housed in each museum, aiding in understanding the spatial dimensions of the museum's art collection.

```python
#What is the average sale price of artworks in each style?
sql_query="Select style,avg(sale_price)as avg_saleprice from work join product_size on work.work_id = product_size.work_id group by style  LIMIT 10"
df_mysql=pd.read_sql(sql_query,connection)
df_mysql
plt.figure(figsize=(10,10))
sns.barplot(data=df_mysql,x='style',y='avg_saleprice',width=0.5)
plt.xlabel('Artwork Styles')
plt.ylabel('AVG Sale Price Artworks in each Styles')
plt.title('AVG Sale Price Artworks in each Styles')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

#  It retrieves the top 10 styles with their corresponding average sale prices. The resulting bar plot visually represents this data, showing the average sale price for each style. This analysis helps in understanding the pricing trends and preferences within the art market, aiding in strategic decision-making for art investments or collection curation.
